<h1>Standar Teleportation Protocol</h1>

In this notebook we will revisit the Standar Teleportation Protocol as propposed initially by Popescu in 1993. 

We star with to agents Alice and Bob that share a two-qubit Maximally Entangled State, let's assume that they have the  Bell state $\ket{\Phi^{+}}$ 

\begin{equation}
\ket{\Phi^{+}}_{AB}= \frac{1}{\sqrt{2}}(\ket{00}+\ket{11})_{AB}
\end{equation}

This for now on will be reffered as the **Resource State**. 

Alice also has another qubit, labeled as $a$, which is in an arbitrary state $\ket{\alpha}_{a}= (\alpha\ket{0}+\beta\ket{1})_{a}$, it's important to recall that they don't need to know the values of $\alpha$ and $\beta$, in fact, this protocol works even if they dont know nothing about this state. The combined state of these three qubits is 

\begin{equation}\begin{split}
\ket{\alpha}_{a}\ket{\Phi^{+}}_{AB}&= \frac{1}{\sqrt{2}} (\alpha\ket{0}+\beta\ket{1})_{a}(\ket{00}+\ket{11})_{AB}\\
\end{split}\end{equation}


Alice's goal is to send this state to Bob. To achieve this, she needs to implement a Bell Measurement on the qubits $a$ and $A$ in her possession. As we learned from the Bell Measurement notebook, this involves applying the gates $H_{a}CNOT(a,A)$. After these operations, the state becomes:

\begin{equation}\begin{split}
H_{a}CNOT(a,A)\ket{\alpha}_{a}&\ket{\Phi^{+}}_{AB}=\frac{1}{\sqrt{2}} H_{a} \big(\alpha\ket{0}_{a} (\ket{00}+\ket{11})_{AB} +\beta\ket{1}_{a} (\ket{10}+\ket{01})_{AB} \big)\\
&=\frac{1}{2} \big(\alpha(\ket{0}+\ket{1})_{a} (\ket{00}+\ket{11})_{AB} +\beta(\ket{0}-\ket{1})_{a} (\ket{10}+\ket{01})_{AB} \big)\\
&=\frac{1}{2} \big(\ket{00}_{aA}(\alpha\ket{0}+\beta\ket{1})_{B}+\ket{01}_{aA} (\beta\ket{0}+\alpha\ket{1})_{B} +\ket{10}_{aA}(\alpha\ket{0}-\beta\ket{1})_{B} + \ket{11}_{aA}(-\beta\ket{0}+\alpha\ket{1})_{B} \big)\\
&=
\frac{1}{2} \big(\ket{00}_{aA} \ket{\alpha}_{B}+\ket{01}_{aA} X \ket{\alpha}_{B} +\ket{10}_{aA} Z\ket{\alpha}_{B} + \ket{11}_{aA} XZ \ket{\alpha}_{B}  \big)\\
\end{split}\end{equation}

In the final line, we can clearly see that Bob's qubit is in a state that's just one local unitary operation away from the state Alice wanted to send.

The protocol concludes when Alice sends the result of her measurement to Bob through classical communication. Based on Alice's measurement result, Bob performs specific operations as follows:


\begin{equation}
\begin{array}{cc}
\text{Alice measure} & \text{Bob operation}\\
\ket{00}_{aA} & \mathbb{I}_{B}\\
\ket{01}_{aA} & X_{B}\\
\ket{10}_{aA} &  Z_{B}\\
\ket{11}_{aA} & ZX_{B}\\
\end{array}
\end{equation}


By following these steps, they ensure that the qubit in Bob's possession ends up in the state Alice initially wanted to send.


This protocol wors perfectly if the resource state is given by the Bell state $\ket{\Phi^{+}}_{AB}$. But what will happend if instead the resource state is the Bell state $\ket{\Psi^{+}}_{AB}$?
